불 인덱싱 조건을 활용하여 특정 기준을 만족하는 행들만 필터링한 부분 표를 생성하였다. 비교 연산자를 통해 1930년 이후의 작품을 추출하였고 복수의 작가를 매칭할 때는 isin 메서드를 사용하여 코드를 간결하게 작성하였다. 이를 통해 조건에 부합하는 작품들만 효율적으로 분리해 낼 수 있음을 확인하였다. 글자수데이터를 판단하여 짧음, 보통, 긴 으로 범주화하는 categorize 함수를 만들었다. 함수를 apply메서드에 사용하여 데이터프레임의 글자수 열에  분량이라는 새로운 파생 열을 생성하였다. 출력 결과를 통해 각 작품의 글자수 크기에 따라 지정한 규칙대로 분량이 정확하게 매핑되었음을 확인하였다.C에서는 sort_values 메서드의 by와 ascending 매개변수에 각각 리스트를 넣어 정렬하였다. 이에 따라 전체 데이터가 일차적으로 '연도' 기준 오름차순 정렬되고 동일한 연도 내에서는 글자수가 많은 순서인 내림차순으로 정렬되었다. 호출 하나만으로 복잡한 정렬 우선순위를 데이터프레임에 올바르게 이용할 수 있음을 확인하였다.
